```python
# MC epsilon greedy
import numpy as np
import random
import GridWorld_v2
import time

start_time = time.time()

gamma = 0.9
rows = 5
columns = 5
trajectorySteps = 200
np.random.seed(521)
gridworld = GridWorld_v2.GridWorld_v2(rows=5, columns=5, forbiddenAreaNums=2, targetNums=1, seed=521, forbiddenAreaScore=-10)
# value = np.zeros(rows * columns)
qtable = np.zeros((rows * columns, 5))
policy = np.eye(5)[np.random.randint(0, 5, size=(rows * columns))]
# print(policy)

gridworld.showPolicy(policy)
cnt = [0 for _ in range(25)]

episode_num = 200
epsilon = 0.1

for episode in range(episode_num):
    if epsilon > 0.001:
        epsilon -= 0.001
    else:
        epsilon = 0.001
    
    prob1 = 1 - epsilon * (4 / 5)
    prob2 = epsilon / 5
    print("trajectorySteps",trajectorySteps)
    print(f"epision:{epsilon}, prob1:{prob1}, prob2:{prob2}")
    
    policy_epsilon = np.zeros_like(policy, dtype=float)
    policy_epsilon[policy == 1] = prob1
    policy_epsilon[policy == 0] = prob2
    i = np.random.randint(0, 24)
    j = np.random.randint(0, 4)

    # cnt = [0 for _ in range(25)]
    qtable_rewards = [[0 for _ in range(5)] for _ in range(rows * columns)] 
    qtable_nums = [[0 for _ in range(5)] for _ in range(rows * columns)]
    Trajectory = gridworld.getTrajectoryScore(i, j, policy_epsilon, trajectorySteps)
    _, _, score, _, _ = Trajectory[trajectorySteps]
    for k in range(trajectorySteps - 1, -1, -1):
        tmpstate, tmpaction, tmpscore, _, _ = Trajectory[k]
        cnt[tmpstate] += 1
        score = tmpscore + gamma * score
        qtable_rewards[tmpstate][tmpaction] += score
        qtable_nums[tmpstate][tmpaction] += 1
        qtable[tmpstate][tmpaction] = qtable_rewards[tmpstate][tmpaction] / qtable_nums[tmpstate][tmpaction]
    
    values = []
    for i in range(25):
        value = 0
        for j in range(5):
            value += policy_epsilon[i][j] * qtable[i][j]
        values.append(value)
    print(np.array(values).mean())
    gridworld.showPolicy(policy)
    policy = np.eye(5)[np.argmax(qtable,axis=1)]
    policy_epsilon = np.zeros_like(policy, dtype=float)
    policy_epsilon[policy == 1] = prob1
    policy_epsilon[policy == 0] = prob2

    # print(np.array(cnt).reshape(5, 5))
print(np.array(cnt).reshape(5, 5))
endtime = time.time()

print(endtime - start_time)
```
